## 1) Use Pandas’ read_csv function directly on this URL to open it as a DataFrame

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "http://www.ldeo.columbia.edu/~danielmw/usgs_earthquakes_2014.csv"
```

```python
df = pd.read_csv(url)

display(df.head())
df.info()
```

## 2) Re-read the data in such a way that all date columns are identified as dates and the earthquake ID is used as the index

```python
df.columns
```

```python
df = pd.read_csv(
    url,
    parse_dates=['time', 'updated'],
    index_col='id'
)
```

```python
display(df.head())
df.info()
```

## 4) Use nlargest to get the top 20 earthquakes by magnitude

```python
top20 = df.nlargest(20, 'mag')
top20[['mag', 'place']]
```

## 5) Extract the state or country using Pandas text data functions

```python
df['country'] = df['place'].str.split(',').str[-1]
display (df[['place', 'country']].head())
df.head()
```

## 6) Display each unique value from the new column

```python
unique_locations = df['country'].unique()
print(unique_locations)
```

## 7) Create a filtered dataset that only has earthquakes of magnitude 4 or larger

```python
filtered_df = df[df['mag'] >= 4]
```

## 8) Using the filtered dataset (magnitude > 4), count the number of earthquakes in each country/state. Make a bar chart of this number for the top 5 locations with the most earthquakes

```python
counts = filtered_df['country'].value_counts()
counts
```

```python
top5 = counts.head(5)
top5
```

```python
top5.plot(kind='bar')
plt.xlabel('Location')
plt.ylabel('Earthquake count')
plt.title('Top 5 Locations with the most Earthquakes (Magnitude ≥ 4)')
plt.show()
```

## 9) Make a histogram the distribution of the Earthquake magnitudes

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 5))

df['mag'].hist(ax=ax1, bins=np.arange(-1,8,0.1), log=True)
ax1.set_title('Unfiltered Magnitudes')
ax1.set_xlabel('Magnitude')
ax1.set_ylabel('Count (Log Scale)')
ax1.set_ylim(1e0, 1e4)

filtered_df['mag'].hist(ax=ax2, bins=np.arange(-1,8,0.1), log=True)
ax2.set_title('Filtered Magnitudes (≥ 4)')
ax2.set_xlabel('Magnitude')
ax2.set_ylabel('Count (Log Scale)')
ax2.set_ylim(1e0, 1e4)
ax2.set_xlim(4,8)

plt.show()
```

Notice a general decrease as you move towards high magnitudes - for both there are more small earthquakes/the highest count is at the lower magnitudes (but not the lowest) and the lower count is at the highest magnitudes

This indicates an inverse relationship where as earthquake magnitude increases, the frequency of earthquakes decreases

Distributions are positively skewed, with tails towards high magnitudes

## 10) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 4))

# Unfiltered
df.plot.scatter(x='longitude', y='latitude', c='mag', colormap='viridis', 
                s=1.5, alpha=0.5, ax=axes[0], title="All Earthquakes", vmin = 0, vmax = 8)

# Filtered
filtered_df.plot.scatter(x='longitude', y='latitude', c='mag', colormap='viridis', 
                         s=1.5, ax=axes[1], title="Magnitude ≥ 4", vmin = 0, vmax = 8)

# Colorbar
cbar1 = axes[0].collections[0].colorbar
cbar2 = axes[1].collections[0].colorbar

cbar1.set_label('Magnitude')
cbar2.set_label('Magnitude')


plt.show()
```

The filtered dataset has less visual cluttering, particularly in the US which was filled with low magnitude earthquake data - this could be overrepresented due to US prevalence of sensors

Seems that smaller earthquakes occur more widely while larger earthquakes are more geographically concentrated

Filtering reviews a more clear structure that fits more with where we know tectonic plates to be
